```python
routes = {
    'Pune' : ['Jaipur','Dehradun'],
    'Jaipur':['Bangalore','Dehradun'],
    'Dehradun':['Mumbai','Surat'],
    'Surat':['Mumbai'],
    'Bangalore':['Dehradun','Hyderabad'],
    'Hyderabad':['Mumbai']
}
```

```python
class Graph:
  def __init__(self,routes):
    self.routes=routes

  def get_paths(self,start,end,path=[]):     #start -> starting point , end -> ending point, path -> An empty list to store all paths.
    path = path+[start]
    if start==end:
      return [path]

    if start not in self.routes:
      return []

    all_paths = []

    for vertex in self.routes[start]:   #vertex will hold all the paths that we can travel from the starting point.
      if vertex not in path:
        new_paths = self.get_paths(vertex,end,path)
        for x in new_paths:  #Whatever path we will get from this new_paths will be stored in all_paths
          all_paths.append(x)
    return all_paths


  def get_min_distance(self, start, end):
        visited = set()
        queue = [(start, 0)]  # (node, distance)

        while queue:
            node, dist = queue[0]
            queue = queue[1:]
            if node == end:
                return dist
            if node not in visited:
                visited.add(node)
                for neighbor in self.routes.get(node, []):
                    queue.append((neighbor, dist + 1))
        return -1
ob = Graph(routes)
ob.get_paths('Pune','Mumbai')
ob.get_min_distance('Pune','Mumbai')
```
