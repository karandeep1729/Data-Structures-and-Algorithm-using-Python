# Linear Search

```python
def Linear_Search(arr,target):
  if len(arr)==0:
      return "Empty Array"
  for x in range(len(arr)):
    if target == arr[x]:
      return "Element found at: ",x

  return -1

# Time complexity of Linear Search => O(N)   N -> Number of data elements present inside our array.
```

```python
Linear_Search([],50)
```

# Binary Search:

```python
# It works only on sorted data.
# It gives us the time complexity of O(log N)
```

```python
def BinarySearch(arr,target):
  if len(arr)==0:
    return "Empty Array"

  start = 0
  end = len(arr)-1
  while start<=end:
    mid = (start+end)//2
    if arr[mid]==target:
      return "Element found at Index" , mid

    elif arr[mid]>target:
      end = mid-1

    else:
      start = mid+1

  return -1
```

```python
arr = [5 , 8 , 10 , 15 , 20 , 25 , 30 , 50 , 55 , 70 , 90]
target = int(input("Enter Target element"))
ans = BinarySearch(arr,target)
print(ans)
```

```python
# Recursive Approach for Binary Search
```

```python
def Binary_Search_Recursive(arr,target,start,end):
  if start > end:
    return -1

  if len(arr) ==0:
    return -1

  mid = (start+end)//2
  if arr[mid]==target:
    return mid

  elif arr[mid]>target:
    return Binary_Search_Recursive(arr,target,start,mid-1)

  else:
    return Binary_Search_Recursive(arr,target,mid+1,end)
```

```python
arr = [5 , 8 , 10 , 15 , 20 , 25 , 30 , 50 , 55 , 70 , 90]
start = 0
end = len(arr)-1
target = -51

Binary_Search_Recursive(arr,target,start,end)
```
