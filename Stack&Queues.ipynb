**Stack** : LIFO (Last In First Out)



1.   *Push Operation* : Used to add a new element on to the top of stack.
2.   *Pop Operation* : Used to remove the top element from the stack.
3.   *Peek Operation* : Used to see the top most element in your stack.



*   **Stack Overflow** : When you try to push a new element when the stack is full.
*   **Stack Underflow** : When you try to remove an element from an empty stack.




# Stacks are used in undo operations, when we run our code functions are stored inside the stack memory.

***Queue***: "First In First Out"

1.   Enqueue Operation: Adding a new element at the end of the queue.
2.   Dequeue Operation: Removing the first element from the queue.
3.   Peek Operation: Retreving the first element from your queue.

```python
from collections import deque
class Stack:
  def __init__(self):
    self.stk = deque()    #self.stk object of deque class.

  def push(self,value):
    self.stk.append(value)

  def pop(self):
    if len(self.stk)==0:
      return "Stack Underflow"
    print("Removing: ",self.stk.pop())

  def peek(self):
    print("Top element on the Stack: ",self.stk[-1])
```

```python
ob = Stack()
ob.push(20)
ob.push(30)
ob.push(40)
ob.pop()
ob.push(50)
ob.push(60)
ob.pop()
ob.pop()
ob.push(27)
ob.push(34)
ob.pop()
ob.pop()
ob.pop()
ob.peek()   #
ob.pop()
ob.pop()
```

```python
class Queue:
  def __init__(self):
    self.q = deque()

  def enqueue(self,value):
    self.q.appendleft(value)

  def dequeue(self):
    if len(self.q)==0:
      return "Empty Queue"
    print(self.q.pop())

  def peek(self):
    return self.q[-1]
```

```python
ob = Queue()
ob.enqueue(10)
ob.enqueue(50)
ob.dequeue()
ob.enqueue(60)
ob.enqueue(20)
ob.dequeue()
ob.enqueue(30)
ob.peek()
```
