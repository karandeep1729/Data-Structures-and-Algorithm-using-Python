```python
# Pre Order Traversal -> Print the value, then go the left child and the go to the right child. ( P L R)
#  This traversal always print the root node at the very first step.

# In Order Traversal -> Go to the left child, print the value and then go to the right child. (L P R)
# This traversal will always print the data in sorted way.

# Post Order Traversal -> Go to the left child , go to the right child and then print the value. (L R P)
# This traversal always print the root node at the last.
```
