```python
# Creating a variable, defining a fxn, printing some values
# takes constant amount of complexity O(1)
# N is used to represent number of data elements in your program.
```

```python
arr = [10,20,30,50,40,60] # -> O(1) -> Constant time complexity
length_arr = len(arr)    # O(1)  -> Constant complexity
for x in range(length_arr):    # This loop is running till the size of my list.  O(N)
  print(arr[x])           # O(1)

# Q: What is the time complexity of this program?
# A: O(1) + O(1) +O(N) + O(1)
# A: O(N+3)  -> O(N)


#  O(N^2+ N) -> O(N^2) -> We willl ignore the smaller terms.
# With respect to N these constant values can be very small. We always ignore the constant while
# calculating complexity. Higher terms are always given prirority.
```

```python
arr = [ 10,13,43,12,11,165,753,34]
length_arr = len(arr)
for x in range(length_arr//2):    # 4 times -> O(Log N)
  print(arr[x])


# Which one is better O(N) or O(LOG N) -> Olog(N) will always take less time than N.
```

```python
def delete_first_element(arr):
  new_arr = []
  N = len(arr)
  for i in range(1,N):      # O(N-1) -> O(N)
    new_arr.append(arr[i])
  return new_arr

# Deleting first element from your array. Time complexity -> O(N)
```

```python
arr = [50,20,30,10,40]
ans = delete_first_element(arr)
print(ans)
ans = delete_first_element(ans)
print(ans)
ans = delete_first_element(ans)
print(ans)
ans = delete_first_element(ans)
print(ans)
ans = delete_first_element(ans)
print(ans)
```
