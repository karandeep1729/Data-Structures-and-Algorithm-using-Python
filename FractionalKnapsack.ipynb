```python
class ItemList:
  def __init__(self,item_name,value,weight):
    self.item_name=item_name
    self.value=value
    self.weight=weight
    self.ratio = value/weight



class Knapsack:
  def __init__(self,bag_capacity):
    self.bag_capacity = bag_capacity

  def sort_items(self,items):
    length_item= len(items)
    for i in range(length_item):
      for j in range(1,length_item):
        if items[j].ratio>items[j-1].ratio:
          temp = items[j]
          items[j] = items[j-1]
          items[j-1] = temp

  def loot(self,items):
    self.sort_items(items)
    total_loot = 0   #Keep track of the total values or profit
    selected_items = []   #Keep track of the items selected.

    for x in items:
      if self.bag_capacity >= x.weight:
        self.bag_capacity = self.bag_capacity - x.weight
        total_loot = total_loot + x.value
        selected_items.append((x.item_name,1))   #1 means that element is completely stored.
      else:
        fraction = self.bag_capacity/x.weight
        total_loot = total_loot + x.value* (self.bag_capacity/x.weight)
        selected_items.append((x.item_name,fraction))
        break

    return [total_loot, selected_items]




```

```python
items = [
    ItemList('Cake',50,10),
    ItemList('Ice Cream',120,20),
    ItemList('Chocolate',150,30),
    ItemList('Brownie',210,40),
    ItemList('Pastry',240,50)
]
```

```python
ob = Knapsack(50)
```

```python
ans = ob.loot(items)
ans
```
