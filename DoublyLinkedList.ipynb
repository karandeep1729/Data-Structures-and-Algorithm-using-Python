```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insertBeg(self, value):
        new_node = Node(value)
        if self.head!=None:
            new_node.next = self.head
            self.head.prev = new_node
        self.head = new_node

    def insertLast(self, value):
        if self.head == None:
            self.insertBeg(value)

        temp = self.head
        while temp.next!=None:
            temp = temp.next

        node = Node(value)
        temp.next = node
        node.prev = temp


    def display(self):
        if self.head == None:
            print("Doubly Linked List is Empty")
            return
        temp = self.head
        while temp is not None:
            print(temp.value, end=" <--> ")
            temp = temp.next
        print("END")

    def display_reverse(self):
        if self.head is None:
            print("Doubly Linked List is Empty")
            return
        temp = self.head

        while temp.next !=None:
            temp = temp.next

        while temp !=None:
            print(temp.value, end=" <--> ")
            temp = temp.prev
        print("END")

# Example usage:
dll = DoublyLinkedList()
dll.insertLast(500)
dll.insertBeg(50)
dll.insertBeg(70)
dll.insertBeg(90)
dll.insertLast(355)
dll.insertBeg(110)
dll.insertBeg(150)

dll.display()          # Display the list in forward direction
dll.display_reverse()  # Display the list in reverse direction
```
