```python
# Create a function which will return the maximum value from an array.
# Input: [50,10,30,20,80]    Output: 4
# Input: [-5,-9,-1,6,-9]     Output: 3
```

```python
def find_Max(arr,last):
  max = 0
  for x in range(1,last+1):
    if arr[max] < arr[x]:
      max =x

  return max

```

# Selection Sort

```python
def SelectionSort(arr):
  last = len(arr)-1
  while last>0:
    maxIndex = find_Max(arr,last)   # Here we will get the index of the largest element in the particular part of array.
    temp = arr[last]
    arr[last] = arr[maxIndex]
    arr[maxIndex] = temp
    last -=1

  return arr

arr = [50,10,20,80,90,100,70,25]
SelectionSort(arr)

# Make less swaps good for larger datasets.
```

# Insertion Sort

image.png

```python
def InsertionSort(arr):
  n = len(arr)-1
  for x in range(n):
    for j in range(x+1,0,-1):
      if arr[j]<arr[j-1]:
        temp = arr[j]
        arr[j] = arr[j-1]
        arr[j-1] = temp
      else:
        break
  return arr
```

```python
arr = [50,10,20,78,94,12,111,-34,235,150,190,120]
InsertionSort(arr)
```

# Merge Sort

```python
def divide_arr(arr):
  if len(arr)<=1:
    return arr

  mid = len(arr)//2
  left = divide_arr(arr[:mid])
  right = divide_arr(arr[mid:])
  return merge(left,right)

def merge(left,right):
  new_arr = []
  i = 0    # For left sub array
  j = 0    # For right sub array

  while i < len(left) and j < len(right):
    if left[i] < right[j]:
      new_arr.append(left[i])
      i+=1
    else:
      new_arr.append(right[j])  #
      j+=1

  #When the upper while loop will be over. All the elements of one array will be stored in new array.
  #But still there will be few elements that might be remaining either of left or right array.


  # If elements are remaining inside left array then this while loop will execute
  while i < len(left):
    new_arr.append(left[i])
    i+=1

  # If elements are remaining inside right array then this loop will execute

  while j < len(right):
    new_arr.append(right[j])
    j+=1


  return new_arr
```

```python
arr = [55,-7,2,0,45,86,7,0.6]
divide_arr(arr)
```

# Quick Sort

image.png

```python
def QuickSort(arr,low,high):
  if low>=high:
    return

  start = low
  end = high
  mid = (start+end)//2
  pivot = arr[mid]

  while start<=end:
    while arr[start]<pivot:  #This loop will stop when there the arr[start]>=pivot element (Violation)
      start+=1
    while arr[end]>pivot:   #This loop will stop when the arr[end] <=pivot element (Violation)
      end-=1

    if start<=end:
      temp = arr[start]
      arr[start] = arr[end]
      arr[end] = temp
      start+=1
      end-=1


  # Now I will divide my array into two segments and sort accordingly.
  # 1st -> from low till end
  # 2nd -> from start till high
  QuickSort(arr,low,end)
  QuickSort(arr,start,high)
```

```python
arr = [ 10 , 5 , 4 , 20 , 14 , 7 , 9 , 12 ]
low = 0
high = len(arr)-1
QuickSort(arr,low,high)
```
