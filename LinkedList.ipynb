```python
class Node:
  def __init__(self,value,next_address=None):
    self.value=value
    self.next_address=next_address


class LinkedList:
  def __init__(self):
    self.head = None
    self.size = 0

  def insertBeg(self,value):  # Time complexity of inserting a new element at the beg will be O(1)
    new_node = Node(value,self.head)    #By passing the address of the head node element will be inserted to the beg of the linkedlist.
    self.head = new_node
    self.size+=1

  # def display_head(self):
  #   print(self.head.value)
  #   print(self.head)


  def insertEnd(self,value):
    if self.head==None:  # This means the linked list is empty we are inserting the very first element.
      self.insertBeg(value)
      return
    temp = self.head
    while temp.next_address!=None:
      temp = temp.next_address

    temp.next_address = Node(value)
    self.size+=1

  def display(self):   # Time complexity will be O(N)
    temp = self.head
    while temp!=None:
      print(temp.value,end="-->")
      temp = temp.next_address
    print("END")

  def deleteMid(self):
    mid = self.size//2
    temp = self.head
    count = 0
    while count<mid-1:  # When this loop will be over temp will be equal to element just before the middle element.
      temp = temp.next_address
      count+=1
    temp.next_address =temp.next_address.next_address
    self.size-=1

ob = LinkedList()
ob.insertEnd(100)
ob.insertBeg(5)
ob.insertBeg(7)
ob.insertBeg(1)
ob.insertEnd(50)
ob.insertBeg(6)
ob.insertBeg(12)
ob.insertEnd(70)
ob.display()
# print(ob.size)
ob.deleteMid()
ob.display()
```
