```python
def recursive_sum(n):
  if n==1:
    return 1
  return n + recursive_sum(n-1)
```

```python
def fibonacci_series(N):  #N = 3
  if N<=0:
    return "Enter a Valid Number"

  elif N==1 or N==2:
    return 1
  return fibonacci_series(N-1) + fibonacci_series(N-2)
              # 1                       # 1
```

```python
fibonacci_series(10)
```
